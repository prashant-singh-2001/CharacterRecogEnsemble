```python
from keras.preprocessing.image import ImageDataGenerator

# Path to your validation dataset
validation_data_dir = '/content/DevanagariHandwrittenCharacterDataset/Test'

# Set up ImageDataGenerator to load and preprocess images
validation_datagen = ImageDataGenerator(rescale=1./255)

# Set the batch size and other parameters
batch_size = 32

# Create a generator for loading validation images from directories
validation_generator = validation_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(32, 32),
    batch_size=batch_size,
    class_mode='categorical',  # Adjust if using different labels
    shuffle=False  # Ensure order is maintained
)

# Extract validation labels
y_val = validation_generator.classes

# y_val contains the labels for the test images
print(y_val)
```

```python
from scipy.stats import mode
from sklearn.metrics import accuracy_score
# Convert the predictions to numpy arrays
all_predictions = np.array(predictions_all)
```

```python
# Combine predictions from all models
combined_predictions = np.concatenate(predictions_all, axis=0)

# Reshape combined predictions to have shape (num_samples, num_models, num_classes)
combined_predictions = combined_predictions.reshape((len(all_predictions[0]), 9, -1))

# Perform majority voting across models
ensemble_predictions = np.argmax(np.apply_along_axis(lambda x: np.bincount(x).argmax(), axis=2, arr=combined_predictions), axis=1)
```

```python
y_val=np.array(y_val)
```

```python
# # Evaluate the performance of the ensemble predictions (e.g., accuracy)
ensemble_accuracy = accuracy_score(y_val, ensemble_predictions)
ensemble_accuracy
```
